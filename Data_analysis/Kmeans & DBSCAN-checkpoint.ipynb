```python
km=KMeans(n_clusters=3)
y_pred=km.fit_predict(df1)
y_pred
```

```python
df1['cluster_kmeans']=y_pred
df1.tail()
```

```python
df2 = df1[df1.cluster_kmeans==0]
df3 = df1[df1.cluster_kmeans==1]
df4 = df1[df1.cluster_kmeans==2]
```

```python
plt.scatter(df2['petal.length'],df2['petal.width'],color='blue')
plt.scatter(df3['petal.length'],df3['petal.width'],color='green')
plt.scatter(df4['petal.length'],df4['petal.width'],color='yellow')
```

```python
sse = []
k_rng = range(1,10)
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df1)
    sse.append(km.inertia_)
```

```python
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
```
